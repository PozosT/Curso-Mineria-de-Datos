
### Parcial II


**Base de Datos**: *StudentPerformanceFactors.csv*  
- **Descripción**: Datos relacionados con los factores que influyen en el rendimiento de los estudiantes.
- **Objetivo**: Predecir si un estudiante alcanzará un buen rendimiento académico (alta puntuación).


### Enunciado:

1. **Conversión de la Variable Objetivo**:
   - Convierte la variable de puntuación en una variable categórica que indique si un estudiante ha obtenido una puntuación "alta" o "baja". Define un umbral adecuado para esta categorización basado en la distribución de las puntuaciones en el conjunto de datos.

2. **Análisis Exploratorio de Datos (EDA)**:
   - Realiza un análisis exploratorio para entender la distribución de las características y su relación con la nueva variable categórica de rendimiento.
   - Identifica patrones, anomalías, y correlaciones que puedan influir en el rendimiento académico de los estudiantes.

3. **Preprocesamiento**:
   - Aplica técnicas de preprocesamiento según lo indicado en el análisis EDA.
   - Considera la codificación de variables categóricas, la estandarización de variables numéricas y el manejo de valores faltantes si es necesario.

4. **Evaluación del Balanceo de Datos**:
   - Evalúa si es necesario aplicar técnicas de balanceo de datos, dada la nueva distribución de la variable categórica de rendimiento.
   - Caso necesario, cual sería la estretegía que aplicarías? No requieres aplicar el código pero si dar la recomendación.


5. **Modelado y Optimización de Hiperparámetros**:
   - **Modelos a entrenar**: Regresión Logística, Regresión Regularizada (Lasso,Ridge y Elasticnet), SVM y Árboles de Decisión.

6. **Selección del Mejor Modelo**:

   - Justifica tu elección explicando por qué consideras que este modelo es el más adecuado para el problema de clasificación y el conjunto de datos.

7. **Evaluación de Modelos**:
   - Evalúa el rendimiento del mejor modelo utilizando el conjunto de testeo y concluye con base en las métricas obtenidas. Evalúa los modelos utilizando métricas como ROC-AUC, precisión, recall, F1-score.
